```python
### Създаване на Series
s = pd.Series([12,-4,7,9,11,-3,1,15,6,2,-9,19,3,5]) # отново представяме стойностите като списък
s1 = pd.Series([10,-5,8,10,12,-4,1,16,7,1,-8,20,4,4])
s # индексите са отляво

#s = pd.Series([12,-4,7,9], index=['a','b','c','d']) # ако искаме индексите да не са по подразбиране, а някакви други
s.values # показва стойностите
s.index # показва индексите
#(s[2], s['b']) # така може да се извикват отделните елементи
#s[3] = 11 # така може да се променят отделни елементи; НЕ може да се добавят нови

s[s > 8] # филтриране на елементите
#np.log(s) # може да се извършват математически операции върху целия ред
s/2

s.drop([1, 4, 6]) # показва Series без указания от индекса елемент. НЕ ИЗТРИВА елементите!

#s.order() # НЕ работи!
s.corr(s1) # корелация между два Series
s.cov(s1) # ковариантност между два Series
```

```python
### Работа с Data Frame
dat = {'color' : ['blue','green','yellow','red','white', 'pink', 'carmine', 'brown', 'clear'],
'object' : ['ball','pen','pencil','paper','mug', 'notebook', 'case', 'box', 'tape'],
'price' : [1.2,1.0,0.6,0.9,1.7, 1.9, 2.0, 1.3, 0.8],
'quantity' : [10, 6, 8, 12, 4, 9, 11, 10, 9]} # дефинираме някакво множество, например чрез dictionary

fra = pd.DataFrame(dat) # дефинираме нашата Data Frame чрез pd.DataFrame(МНОЖЕСТВО)
fra

fra2 = pd.DataFrame(dat, columns=['object','price']) # дефинираме друга Data Frame, съдържаща тези колони
fra2

# бързо генериране на пример и поставяне на имена на колони и индекси(редове)
fra3 = pd.DataFrame(np.arange(25).reshape((5,5)), index=['red','blue','yellow','white', 'purple'],columns=['ball','pen','pencil','paper','mug'])
fra3

fra4 = pd.DataFrame([[1,4,3,6,7],[4,5,6,1,3],[3,3,1,5,2],[4,1,6,4,5],[2,5,3,6,1]], index=['red','blue','yellow','white', 'purple'],columns=['ball','pen','pencil','paper','mug'])

fra.columns # показва имената на колоните
#fra.index # показва имената на редовете, т.е. индексите
#fra.values # показва какво има в таблицата

fra['object'] # показва съдържанието на конкретна колона
fra.object # показва съдържанието на конкретна колона

fra[['color','object']] # показва само тези колони от DaFr
fra[fra.columns[1:4]] # показва само тези колони от DaFr

fra[2:3] # извиква определени редове
#fra['object'][2] # показва конкретен елемент от дадена колона и индекс
```

```python
#### Функции върху колона или ред. Те ще работят върху array и дават един единствен резултат

#f = lambda x: x.max() - x.min() # дефинираме функция, която намира range на данните в колона x

def f(x): # функцията може да се дефинира и така. Прилагаме функцията с ИМЕ–ДАННИ.apply(f)
    return x.max() - x.min()

fra3.apply(f) # получаваме Series от по едно число за колона; за да приложим върху ред, трябва да уточним axis=1
#fra3.apply(f, axis=1) # прилагаме функцията f за всеки ред

def f1(x): # може да се върне Series, а не един скалар
    return pd.Series([x.min(), x.max()], index=['min','max'])

fra3.apply(f1) # получаваме Data Frame
fra3
```

```python
#### Статистически функции
fra3.sum()
fra3.mean()
fra3.describe() # систематизирана таблица с данни за дадената Data Frame

fra3.sort_index(ascending=False) # сортира по индекса на реда: големина или азбуч. ред, by default .sort_index() е нарастващ
fra3.sort_index(axis=1) # сортира колони, т.е. размества колоните по азбучен ред

#fra3.sort_index(by='pen') # сортира по стойностите в определена колона. НЕ РАБОТИ ТУК!
#fra3.dropna(how='all') # премахва NaN, но само ако целият ред е с NaN
fra3.fillna(0) # замества NaN с избрано число, напр. 0
fra3.fillna({'ball':1,'mug':0,'pen':99}) # замества NaN в поцочените колони с избрано число

fra3.cov() # covariance и correlation за Data Frame
fra4.corr() # резултатът е Data Frame
fra3.corrwith(fra4) # pairwise correlation с друга матрица или Series
```
