```python
t = np.arange(0,2.5,0.1) # първо дефинираме стойност по x оста, като използваме np.arange(начало, край, стъпка)
y1 = np.sin(np.pi*t) # след това дефинираме в явен вид какво ще има по y
y2 = np.sin(np.pi*t+np.pi/2)
y3 = np.sin(np.pi*t-np.pi/2)
plt.plot(t,y1,'b--',t,y2,'g',t,y3,'r-.') # накрая подаваме към plt.plot()
```

```python
### Разделяне на хоризонтални subplots
t = np.arange(0,5,0.1) # начало, край, стъпка
y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)

plt.subplot(211) # разделяне на subplots (брой по вертикала, брой по хоризонтала, за кой subplot са командите)
plt.plot(t,y1,'b-.')

plt.subplot(212)
plt.plot(t,y2,'r--')
```

```python
### Разделяне на вертикални subplots
t = np.arange(0.,2,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)

plt.subplot(121)
plt.plot(t,y1,'b-.')

plt.subplot(122)
plt.plot(t,y2,'r--')
```

```python
### Разделяне на вертикални и хоризонтални subplots
t = np.arange(0.,2,0.05)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)
y3 = y1**2
y4 = y2**2

plt.subplot(221)
plt.plot(t,y1,'b-.')

plt.subplot(222)
plt.plot(t,y2,'r--')

plt.subplot(223)
plt.plot(t,y3,'g--')

plt.subplot(224)
plt.plot(t,y4,'g-.')
```

```python
##### Line Chart
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(3*x)/x
y2 = np.sin(2*x)/x
y3 = np.sin(x)/x
plt.plot(x,y,'m--')
plt.plot(x,y2)
plt.plot(x,y3)
#  ако искаме по оста x да пише не числа, а pi. [списък с положенията на ticks],[списък с означенията на ticks]
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],[r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3],[r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])

# ако искаме координатните оси да минават през координатното начало
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

# ако искаме да добавим текст и стрелка към дадена точка от графиката
plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}= 1$', xy=[0,1],xycoords='data',
             xytext=[30,30],fontsize=16,textcoords='offset points',arrowprops=dict(arrowstyle="->",
                                                                                   connectionstyle="arc3,rad=.2"))
```

```python
##### Bar Chart: по оста x имаме категории, а не числа
#index = [0,1,2,3,4]
#index = np.arange(5)
#values1 = [5,7,3,4,6]
#plt.bar(index,values1)
#plt.xticks(index+0.4,['A','B','C','D','E']) #(списък с координати, списък с категории)

# Bar Chart с означение на грешката (стандартното отклонение)
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('УРАААААА!')
plt.bar(index,values1,yerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='First')
plt.xticks(index-0.07,['A','B','C','D','E'])
plt.legend(loc=2)
```

```python
#### Horizontal Bar Chart
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('A Horizontal Bar Chart')
plt.barh(index,values1,xerr=std1,error_kw={'ecolor':'0.1','capsize':6},alpha=0.7,label='First')
plt.yticks(index+0.4,['A','B','C','D','E'])
plt.legend(loc=5)
```

```python
#### Multiseries Bar Chart. Може също да е хоризонтална
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,5,0,8])
plt.title('A Multiseries Bar Chart',fontsize=20)
plt.bar(index,values1,bw,color='b')
plt.bar(index+bw,values2,bw,color='c')
plt.bar(index+2*bw,values3,bw,color='m')
plt.xticks(index+1.5*bw,['A','B','C','D','E'])
```

```python
#### Multiseries Bar Chart with pandas DataFrame. Може да е хоризонтална
data = {'series1':[1,3,4,3,5],'series2':[2,4,5,2,4],'series3':[3,2,3,1,3]}
df = pd.DataFrame(data)
df.plot(kind='bar')
```

```python
#### Multiseries Stacked Bar Charts. Може да е хоризонтална
series1 = np.array([3,4,5,3])
series2 = np.array([1,2,2,5])
series3 = np.array([2,3,3,4])
index = np.arange(4)
plt.axis([-0.4,4,0,15])
plt.bar(index,series1,color='r')
plt.bar(index,series2,color='b',bottom=series1)
plt.bar(index,series3,color='g',bottom=(series2+series1))
plt.xticks(index-0.01,['Jan15','Feb15','Mar15','Apr15'])
```

```python
#### Stacked Bar Charts with pandas DataFrame
data = {'series1':[1,3,4,3,5],'series2':[2,4,5,2,4],'series3':[3,2,3,1,3]}
df = pd.DataFrame(data)
df.plot(kind='bar', stacked=True)
```

```python
#### Other Bar Chart Representations: Mirror
x0 = np.arange(8)
y1 = np.array([1,3,4,6,4,3,2,1])
y2 = np.array([1,2,5,4,3,3,2,1])
plt.ylim(-7,7)
plt.bar(x0,y1,0.9,facecolor='r',edgecolor='w')
plt.bar(x0,-y2,0.9,facecolor='b',edgecolor='w')
plt.xticks(())
plt.grid(True)

for x, y in zip(x0, y1):
    plt.text(x + 0.4, y + 0.05, '%d' % y, ha='center', va= 'bottom')

for x, y in zip(x0, y2):
    plt.text(x + 0.4, -y - 0.05, '%d' % y, ha='center', va= 'top')
```

```python
#### Advanced Plots. Contour plot
dx = 0.01; dy = 0.01
x = np.arange(-np.pi,np.pi,dx) # оригинално тук беше -2.0, 2.0
y = np.arange(-np.pi,np.pi,dy) # оригинално тук беше -2.0, 2.0
X,Y = np.meshgrid(x,y)

def f(x,y):
    return (2-np.cos(x) - np.cos(y))
#    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)
#    return ( y**2 - x**2)

fig, ax = plt.subplots(figsize=(8, 8))  # set figsize to (width, height) in inches

C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8)
plt.clabel(C, inline=1, fontsize=10)

# # може да се добавят още детайли.
# dx = 0.01; dy = 0.01
# x = np.arange(-2.0,2.0,dx)
# y = np.arange(-2.0,2.0,dy)
# X,Y = np.meshgrid(x,y)

# C = plt.contour(X,Y,f(X,Y),8,colors='black')
# plt.contourf(X,Y,f(X,Y),8,cmap=plt.cm.hot)
# plt.clabel(C, inline=1, fontsize=10)
# plt.colorbar()
```
